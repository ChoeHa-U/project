```python
import numpy as np
import pandas as pd
from google.colab import files
uploaded = files.upload()
cars = pd.read_csv("1774404110.csv", na_values="?")
print(cars.shape)
```

```python
# NumPy Problem 1
fuel = cars["mpg"].to_numpy()
print("Mean   :", np.mean(fuel))
print("Median :", np.median(fuel))
print("Std Dev:", np.std(fuel))
print("Above 25 mpg:", np.sum(fuel > 25))

# NumPy Problem 2
cyl_count = cars["cylinders"].to_numpy()
names = cars["car name"].to_numpy()
heavy_cars = names[cyl_count > 6].tolist()
print(heavy_cars)

# NumPy Problem 3
power = cars["horsepower"].to_numpy(dtype=float)
power[np.isnan(power)] = np.nanmean(power)
boosted = power * 1.10
print(boosted[:5])

# NumPy Problem 4
engine_size = cars["displacement"].to_numpy(dtype=float)
made_in = cars["origin"].to_numpy()
print(np.mean(engine_size[made_in == 2]))

# NumPy Problem 5
for c in np.unique(cyl_count):
    print(f"{c} cylinders -> avg mpg: {np.mean(fuel[cyl_count == c]):.2f}")
```

```python
# Pandas Problem 1
print(cars.head(10))
print(cars.shape)
print(cars.describe())

# Pandas Problem 2
yr75_light = cars[(cars["model year"] == 75) & (cars["weight"] < 3000)]
print(yr75_light[["car name", "weight", "mpg"]])

# Pandas Problem 3
print(cars.isnull().sum())
print("Any missing?", cars.isnull().any().any())
cars["horsepower"] = cars["horsepower"].fillna(cars["horsepower"].median())
print("Still missing:", cars["horsepower"].isnull().sum())

# Pandas Problem 4
cars["power_to_weight_ratio"] = cars["horsepower"] / cars["weight"]
print(cars[["car name", "power_to_weight_ratio"]].head())

# Pandas Problem 5
print(cars.groupby("origin")["mpg"].mean())

# Pandas Problem 6
yearly = cars.groupby("model year")[["mpg", "weight", "horsepower"]].mean().round(2)
print(yearly)

# Pandas Problem 7
sippers = cars[cars["mpg"] > 30][["mpg", "cylinders", "horsepower", "weight"]]
sippers.to_csv("high_mpg_cars.csv", index=False)
print("Saved! Total rows:", len(sippers))
```
